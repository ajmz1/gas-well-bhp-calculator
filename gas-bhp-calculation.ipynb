#### Mechanical Energy Balance
Recalling the mechanical energy balance for steady-state flow in a conduit:

$\cfrac{144}{\rho}dp + \cfrac{g}{g_c}dZ + \cfrac{v}{g_c}dv +dF = -dw_s$

This can be simplified for the case of no shaft work and a static pressure column. Additionally, it is assumed that the local gravitational acceleration equals the gravitational acceleration constant, $g = g_c$:

$\cfrac{144}{\rho}dp + dZ = 0$

$dp = - \cfrac{\rho}{144}dZ$

#### Definition of Elevation Gradient (Slant Wells)
$\cos\theta = \cfrac{Z}{L}$

where,

$Z = \text{Depth of well}$

$L = \text{Total well length}$

In differential form:

$\cos\theta = \cfrac{dZ}{dL}$

#### Fluid Properties of Real Gases
$\rho_g = \cfrac{pM}{zRT} = \cfrac{28.97 \gamma_g p}{zRT}$

```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt

# Abou-Kassem and Dranchuk Z-Factor Correlation Function
def z_factor_correlation(Ppr, Tpr, x0= 0.9, x1= 0.8):
    sol = root_scalar(optimization_function, x0= x0, x1= x1, args= (Ppr, Tpr))
    root_estimate = sol.root
    z_estimate = 0.2700 * Ppr / (root_estimate * Tpr)
    return z_estimate

# Optimization function for finding the root of the residual function (default is secant method)
def optimization_function(y, Ppr, Tpr):
    a_arr = np.array([0.3265, -1.0700, -0.5339, 0.01569, -0.05165, 0.5475, -0.7361, 0.1844, 0.1056, 0.6134, 0.7210])
    R1 = a_arr[0] + a_arr[1] / Tpr + a_arr[2] / Tpr**3 + a_arr[3] / Tpr**4 + a_arr[4] / Tpr**5
    R2 = 0.27 * Ppr / Tpr
    R3 = a_arr[5] + a_arr[6] / Tpr + a_arr[7] / Tpr**2
    R4 = a_arr[8] * (a_arr[6] / Tpr + a_arr[7] / Tpr**2)
    R5 = a_arr[9] / Tpr**3
    
    residual_function = (R5 * y**2 * (1 + a_arr[10] * y**2) * np.exp(-1.0 * a_arr[10] * y**2)) + R1 * y - R2 / y \
                        + R3 * y**2 - R4 * y**5 + 1.0
    return residual_function

# Test case showing z-factor for {p_pr, T_pr} = 4.42, 1.47
z_factor_correlation(4.42, 1.47)
z_factor_correlation(4.62668666, 1.21)
```

#### Calculating Bottomhole Static Pressure with Average Temperature and z-Factor Method
>Input Parameter Description and Units:
>
>- specific_gravity ------------> Gas specific gravity, dimensionless
>- pressure_pc ----------------> Pseudocritical pressure, psia
>- temperature_pc -----------> Pseudocritical temperature, degreeRankine
>- pressure_surface ---------> Surface tubing pressure, psia
>- temperature_surface ----> Surface temperature, degreeFahrenheit
>- temperature_bh ----------> Bottomhole temperature, degreeFahrenheit
>- L_well -----------------------> Total length of well, feet
>- slant_angle ----------------> Deviation angle from vertical, degrees



```python
# Problem Input Parameters (Ex. 4.1)
specific_gravity = 0.75
pressure_pc = 667
temperature_pc = 408
pressure_surface = 2500
temperature_surface = 35
temperature_bh = 245
L_well = 10000
slant_angle = 0
input_parameters = np.array([specific_gravity, pressure_pc, temperature_pc, pressure_surface,
                            temperature_surface, temperature_bh, L_well, slant_angle])

def average_temp_and_pressure_method(inputs):
    # Unpack input parameters
    specific_gravity = inputs[0]
    pressure_pc = inputs[1]
    temperature_pc = inputs[2]
    pressure_surface = inputs[3]
    temperature_surface = inputs[4]
    temperature_bh = inputs[5]
    L_well = inputs[6]
    slant_angle = inputs[7]
    
    # Convert surface and bottomhole temperatures
    temperature_surface = degreeF_to_degreeR(temperature_surface)
    temperature_bh = degreeF_to_degreeR(temperature_bh)
    
    # Initial pressure guess and average temperature
    pws_initial = pressure_surface + 1/4 * (pressure_surface / 100) * (L_well * np.cos(slant_angle) / 100)
    temperature_average = np.mean((temperature_surface, temperature_bh))
    temperature_pr = temperature_average / temperature_pc
    
    # Begin iteration loop until pressure convergence
    iteration, max_iteration, pressure_tolerance, pressure_norm = 0, 1000, 0.01, 0.02
    pressure_old = pws_initial
    while pressure_norm > pressure_tolerance and iteration < max_iteration:
        iteration += 1
        
        # Compute pseudo-reduced pressure and z-factor
        pressure_average = np.mean((pressure_surface, pressure_old))
        pressure_pr = pressure_average / pressure_pc
        z_average = z_factor_correlation(pressure_pr, temperature_pr)

        # Update the estimate for bottomhole pressure
        pressure_new = compute_pressure_update(pressure_surface, specific_gravity, L_well, slant_angle, z_average,
                                              temperature_average)
        
        # Compute pressure norm for convergence check
        pressure_norm = np.linalg.norm(pressure_new - pressure_old)

        if pressure_norm > pressure_tolerance and iteration < max_iteration:
            pressure_old = pressure_new
    
    estimated_pressure_bh = pressure_new
    return estimated_pressure_bh, print(f'Estimated Bottomhole Pressure: {estimated_pressure_bh} psia')

# Auxiliary functions
def compute_pressure_update(pressure_surface, specific_gravity, L_well, slant_angle, z_average, temperature_average):
    numerator = 0.01875 * specific_gravity * L_well * np.cos(slant_angle)
    denominator = z_average * temperature_average
    pressure_update = pressure_surface * np.exp(numerator / denominator)
    return pressure_update

def degreeF_to_degreeR(temperature_in_degreeF):
    return temperature_in_degreeF + 459.67

result = average_temp_and_pressure_method(input_parameters)

fig, ax = plt.subplots(dpi= 150)
ax.plot([pressure_surface, result[0]], [0, L_well*np.cos(slant_angle)])
ax.invert_yaxis()
ax.set_xlabel('Presure in the well [psia]')
ax.set_ylabel('True vertical depth [feet]')
ax.grid(which= 'major', axis= 'both')
plt.title('Static Bottomhole Pressure Gradient with No Discretizations')
plt.show()
```
